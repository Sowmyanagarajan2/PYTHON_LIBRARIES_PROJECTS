```python
#data collection -> anyformat(xml,xlxs,csv,db,xml)
#preprocessing -> data end - to end
# model training
# deployment
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Name':['Tom', 'nick', 'krish', 'jack'],
    'Math':[50, 40, 76, 95],
    'Science':[35, 86, 37, 21],
    'English':[23, 53, 33, 12],
    'Commerce':[45, 76, 34, 12],
}
df = pd.DataFrame(data)
# print(df)
#convert to csv
df.to_csv('data.csv', index=False)
#convert to excel
df.to_excel('data.xlsx', index=False)

#read
df = pd.read_csv('data.csv')
# print(df)
df = pd.read_excel('data.xlsx')
# print(df)
#calculate avg marks
df['avg'] = df[['Math', 'Science', 'English', 'Commerce']].mean(axis=1)
# print(df)
#calculate total marks
df['total'] = df[['Math', 'Science', 'English', 'Commerce']].sum(axis=1)
# print(df)
print(df)
topper = df.loc[df['avg'].idxmax()]
print('\n Topper ',topper)
subject_avg = df[['Math', 'Science', 'English', 'Commerce']].mean()
print('\n Subject wise avg ',subject_avg)

#plot
df.plot(kind='bar', x='Name', y=['Math', 'Science', 'English', 'Commerce'])
plt.show()

subject_avg.plot(kind="bar",color=['blue','red','green','black'])
plt.title("Subject wise avg")
plt.xlabel("Subjects")
plt.ylabel("Avg marks")
plt.legend()
plt.show()

plt.scatter(df['Math'], df['Science'])
plt.title("Math vs Science")
plt.xlabel("Math")
plt.ylabel("Science")
plt.show()

#heatmap
# Select only the numerical columns for correlation calculation
numerical_df = df[['Math', 'Science', 'English', 'Commerce', 'avg', 'total']]
sns.heatmap(numerical_df.corr(), annot=True)
plt.show()
```
